# Исследование закономерностей определяющих успешность игр в интернет-магазине «Стримчик»

***Описание проекта:***

Интернет-магазин «Стримчик» - поставщик компьютерных игр по всему миру. Из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation). Педоставлены данные до 2016 года, в 2017 планируется рекламаня кампания. 

В наборе данных попадается аббревиатура ESRB (Entertainment Software Rating Board) — это ассоциация, определяющая возрастной рейтинг компьютерных игр. ESRB оценивает игровой контент и присваивает ему подходящую возрастную категорию, например, «Для взрослых», «Для детей младшего возраста» или «Для подростков».

***Цели исследования:***
- Выявление определяющие успешность игры закономерностей 
- Выявление потенциально популярного продукта
- Отработатка принципа работы с данными
- Проверка гипотез

***Гипотезы:***
- Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;
- Средние пользовательские рейтинги жанров Action и Sports разные.

***Данные:***

<code>Name</code>  — название игры <br>
<code>Platform</code> — платформа<br>
<code>Year_of_Release</code> — год выпуска<br>
<code>Genre</code> — жанр игры<br>
<code>NA_sales</code> — продажи в Северной Америке (миллионы проданных копий)<br>
<code>EU_sales</code> — продажи в Европе (миллионы проданных копий)<br>
<code>JP_sales</code> — продажи в Японии (миллионы проданных копий)<br>
<code>Other_sales</code> — продажи в других странах (миллионы проданных копий)<br>
<code>Critic_Score</code> — оценка критиков (максимум 100)<br>
<code>User_Score</code> — оценка пользователей (максимум 10)<br>
<code>Rating</code> — рейтинг от организации ESRB<br>

##### Причины пропусков:
К сожалению, нет информации о том, как данные были получены, поэтому есть несколько вариантов причины для всех пропусков:
1. Неполадка в системе учета данных, технические баги или человеческий фактор
2. Отсутсвие информации, например у игры может не быть рейтинга(ERSB появился только в 1994 году). Также невозможность установить оценки пользователей до развития глобализации и интернета. Релиз игры часто переносится и установить точную дату выпуска невозможно. В дополнение, некоторые игры могут быть не популярными и данные по ним просто не собирают.

##### Удаление пропусков:

Обработка:
1. name - удаляем, т.к. невозможно заменить пропуски,  а строк не так много.
2. year_of_release - проверим, есть ли год у игр на других платформах, если да, то заменим на него. RP(Rating Pending)
4. critic_score - меняем на 0
5. user_score - меняем на 0

Оставшиеся пропуски оставляем, т.к. слишком много, а данные из других столбцов могут быть полезными. Изменение данных может повлиять на вывод. 

```python
#1
data = data.dropna(subset = ['name'])
```

Проверка на наличие данных у других платформ:

```python
#2
(data[data['year_of_release'].isnull() == True]).head()
```

```python
data[data['name'] == 'Madden NFL 2004']
```

Для платформы PS2 года выпуска нет, но есть на других платформах. Значит можем заменить пропуски значениями из других строк.

```python
data['year_of_release'] = data.groupby('name')['year_of_release'].fillna(method='ffill')
data['year_of_release'] = data.groupby('name')['year_of_release'].fillna(method='bfill')
data = data.dropna(subset = ['year_of_release'])
```

Делаем тоже самое для rating, critic_score и user_score

```python
#3
data['rating'] = data.groupby('name')['rating'].fillna(method='ffill')
data['rating'] = data.groupby('name')['rating'].fillna(method='bfill')
data['rating'] = data['rating'].fillna('RP')
```

#### Анализ рынка видеоигр с 1980 по 2016

```python
name_pivot = data.pivot_table(index = 'name', values = 'total_sales').reset_index()
year_pivot =  data.pivot_table(index = 'year_of_release', values = 'total_sales', aggfunc = ['count', 'sum']).reset_index()
year_pivot.columns = ['year_of_release', 'count', 'sum']
print('Анализ рынка видеоигр:')
print()
print('Игры начали выпускать в', data['year_of_release'].min(), 'году')
print('Крайняя игра была выпущена в', data['year_of_release'].max(), 'году')
print('С 1980 было выпущено', len(data['name'].unique()), 'игр')
print('За этот же период было сделано', len(data['name']),'релизов на разных платформах')

print('Пик активности был в', (year_pivot.loc[year_pivot['count'] == year_pivot['count']
                                               .max(), 'year_of_release']).max(), 'году.')

print('В этот год было выпущено', data['year_of_release'].value_counts().max(), 'игр')

print('А заработано', year_pivot['sum'].max().round(2), 'млн. долларов')

print('Самая популярная игра', (name_pivot.loc[name_pivot['total_sales'] == name_pivot['total_sales'].max(), 'name']).max())
print('Она заработала', name_pivot['total_sales'].max().round(2), 'млн. долларов')
print()
data.pivot_table(index = 'year_of_release', values = 'name', aggfunc = 'count').plot(kind = 'bar', figsize=(15,5))
plt.title('График по количеству игр выпускаемые в разные годы')
```

- С 1980 по 2000 игровая индустрия была еще не освоена и плохо развита, поэтому кол-во игр в этот период имеет минимальные значения.

- Самые продуктивные года были с 2006 по 2010, в этот период выпускалось максимальное кол-во игр.
- С 2010 индустрия идет на спад, в 2016 достигнув минимума с 2001 года. 

Скорее всего, спад после 2008 связан с глобальным кризисом, после которого практически все индустрии пошли на спад, а также снизилась покупательная способность. Это подтвержает график продаж ниже.

#### Корреляция

Создадим функцию, чтобы применить ее ко всем топ5 платформам

```python
def correlation(p):
    print('Анализ зависимомости продаж и оценок в платформе', p)
    print()
    user_cor = new_data[new_data['platform']==p]['user_score'].corr(new_data[new_data['platform']==p]['total_sales']).round(4)
    if 0 <= abs(user_cor) <= 0.3:
        print('Корреляция между продажами и оценками пользователей слабая и равна:', user_cor.round(4))
    elif 0.3 < abs(user_cor) <= 0.7:
        print('Корреляция между продажами и оценками пользователей средняя и равна:', user_cor.round(4))
    elif 0.7 < abs(user_cor) <= 1.0:
        print('Корреляция между продажами и оценками пользователей сильная и и равна:', user_cor.round(4))
        
    new_data[new_data['platform']==p].plot(x='user_score', y='total_sales', kind='scatter', alpha=0.8, figsize=(15,5))
    plt.title('Корреляция продаж и оценок пользователей')
    
    critic_cor = new_data[new_data['platform']==p]['critic_score'].corr(new_data[new_data['platform']==p]['total_sales']).round(4)
    if 0 <= abs(critic_cor) <= 0.3:
        print('Корреляция между продажами и оценками критиков слабая и равна:', critic_cor.round(4))
    elif 0.3 < abs(critic_cor) <= 0.7:
        print('Корреляция между продажами и оценками критиков средняя и равна:', critic_cor.round(4))
    elif 0.7 < abs(critic_cor) <= 1.0:
        print('Корреляция между продажами и оценками критиков сильная и и равна:', critic_cor.round(4))
        
    new_data[new_data['platform']==p].plot(x='critic_score', y='total_sales', kind='scatter', alpha=0.8, figsize=(15,5))
    plt.title('Корреляция продаж и оценок критиков')
```

### Проверка гипотез

#### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

H0: 
Средние пользовательские рейтинги платформ Xbox One = Средние пользовательские рейтинги платформ PC

H1: 
Средние пользовательские рейтинги платформ Xbox One ≠ Средние пользовательские рейтинги платформ PC

```python
xbox_one = data_from_2013.loc[data_from_2013['platform'] == 'XOne','user_score'].dropna()
pc = data_from_2013.loc[data_from_2013['platform'] == 'PC','user_score'].dropna()
```

```python
alpha = 0.05

results = st.ttest_ind(xbox_one, pc, equal_var=False)

print('p-значение:', results.pvalue)


if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Нулевая гипотеза не отвергнута.  Вероятность получить одинакове средние рейнтинги по двум платформам есть.

#### Средние пользовательские рейтинги жанров Action и Sports разные.

H0: 
Средние пользовательские рейтинги жанров Action = Средние пользовательские рейтинги жанров Sports

H1: 
Средние пользовательские рейтинги жанров Action ≠ Средние пользовательские рейтинги жанров Sports

```python
action = data_from_2013.loc[data_from_2013['genre'] == 'Action','user_score'].dropna()
sports = data_from_2013.loc[data_from_2013['genre'] == 'Sports','user_score'].dropna()
```

```python
alpha = 0.05

results = st.ttest_ind(action, sports, equal_var=False)

print('p-значение:', results.pvalue)


if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Нулевая гипотеза отвергнута. Вероятности получить одинакове средние рейнтинги по двум жанрам нет.
